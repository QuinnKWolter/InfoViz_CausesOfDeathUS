```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
%matplotlib inline
mpl.style.use('ggplot')
```

```python
google_df = pd.read_csv('final_Google_by_year.csv', index_col=0)
google_df.head(10)
```

```python
nchs_df = pd.read_csv('final_NCHS_by_year.csv', index_col=0)
nchs_df.head(10)
```

```python
years = list(range(2004, 2018))
columns = ['Heart', 'Cancer', 'CLRD', 'Accident', 'Stroke', 'Alzheimers', 'Diabetes', 'Flu', 'Kidney', 'Suicide']
google_df = google_df.reindex(columns=columns)
google_df['Year'] = years
nchs_df = nchs_df.reindex(columns=columns)
nchs_df['Year'] = years
colors = ['firebrick', 'lightsalmon', 'khaki', 'forestgreen', 'turquoise', 'dodgerblue', 'indigo', 'violet', 'saddlebrown', 'black']

fig, (ax, ax2) = plt.subplots(ncols=2, sharey=True)
google_df.plot(x='Year', kind='bar', stacked=True, color=colors, ax = ax, figsize=(15,7), legend=False)
nchs_df.plot(x='Year', kind='bar', stacked=True, color=colors, ax = ax2, figsize=(15,7))
ax2.legend(bbox_to_anchor=(-0.1,1.3), loc='upper center', ncol=5, prop={'size': 14})
ax.get_yaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)
ax.set_title('What Americans Search For\n(Relative Share of Google Search Interest)', fontsize=19)
ax2.set_title('What Americans Die Of\n(Relative Share of Leading Causes of Death)', fontsize=19)
ax.set(xlabel=None)
ax2.set(xlabel=None)
plt.rc('xtick', labelsize=20) 

plt.show()
```
